```python
import os
import pickle as pkl

from pydeseq2.dds import DeseqDataSet
from pydeseq2.default_inference import DefaultInference
from pydeseq2.ds import DeseqStats
from pydeseq2.utils import load_example_data
import pandas as pd

data = pd.read_excel('C:\\Users\\yylqy\\Desktop\\BMC_Revise\\Puptative_VG\\Tdro\\Tdro.vg.ca.counts.xlsx',index_col=0,header=0)
data = data.T
data = data.applymap(lambda x : round(x))
data
```

```python
meta_data = pd.DataFrame(['vg','vg','vg','ca','ca','ca'],columns = ['condition'])
meta_data.index  = ['vg1','vg2','vg3','vgca1','vgca2','vgca3']
meta_data
```

```python
counts_df = data
genes_to_keep = counts_df.columns[counts_df.sum(axis=0) >= 6]
counts_df = counts_df[genes_to_keep]
inference = DefaultInference(n_cpus=8)
dds = DeseqDataSet(
    counts=counts_df,
    metadata=meta_data,
    design_factors="condition",
    refit_cooks=True,
    inference=inference,
)
dds
```

```python
dds.deseq2()
```

```python
print(dds.varm["LFC"])
```

```python
stat_res = DeseqStats(dds, inference=inference)
stat_res.results_df.to_csv('vg.vgca.deseq2.csv' )
```
